```python
%%writefile app.py
import streamlit as st
import google.generativeai as genai
from PIL import Image
import io

# Configure the Gemini API with your API key
genai.configure(api_key="AIzaSyCO-gMMaUL5__xnDr8do1FXGRaqMOAYdI8")  # Replace with your actual API key

# Initialize the GenerativeModel
model = genai.GenerativeModel('gemini-1.5-flash')  # Choose the appropriate model

# Function to generate text response
def generate_text_response(prompt):
    response = model.generate_text(prompt=prompt)
    return response.result

# Function to process image
def process_image(image_data):
    image = Image.open(io.BytesIO(image_data))
    response = model.generate_content([image])
    return response.text

# Function to process text and image
def process_text_image(image_data, text_prompt):
    image = Image.open(io.BytesIO(image_data))
    response = model.generate_content([text_prompt, image])
    return response.text

# Streamlit App
st.title("Gemini LLM Multi-Modal AI Hands-On")
st.subheader("Bring Your Own Dataset (Kaggle, Local, APIs)")

# Text Processing
st.write("## Text Processing")
text_input = st.text_area("Enter a prompt:")
if st.button("Generate Text Response"):
    if text_input:
        st.write(generate_text_response(text_input))
    else:
        st.warning("Please enter a text prompt.")

# Image Processing
st.write("## Image Processing")
uploaded_file = st.file_uploader("Upload an Image", type=["jpg", "png", "jpeg"])
if uploaded_file:
    st.image(uploaded_file, caption="Uploaded Image", use_column_width=True)
    if st.button("Analyze Image"):
        image_data = uploaded_file.read()
        st.write(process_image(image_data))

# Text + Image Processing
st.write("## Text + Image Processing")
text_prompt_2 = st.text_area("Enter a text prompt for the image:")
if uploaded_file and text_prompt_2:
    if st.button("Analyze Text + Image"):
        image_data = uploaded_file.read()
        st.write(process_text_image(image_data, text_prompt_2))
    else:
        st.warning("Please enter a text prompt and upload an image.")

# Kaggle Dataset Integration
st.write("## Load Kaggle Dataset")
dataset_name = st.text_input("Enter Kaggle dataset name:")
if st.button("Download Dataset"):
    if dataset_name:
        import os
        os.system(f"kaggle datasets download -d {dataset_name} -p ./data && unzip ./data/{dataset_name}.zip -d ./data")
        st.write(f"Downloaded {dataset_name} successfully!")
    else:
        st.warning("Please enter a Kaggle dataset name.")
```
